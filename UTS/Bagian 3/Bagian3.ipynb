```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
# Mount Google Drive
drive.mount('/content/drive')

# Muat dataset
data = pd.read_csv('/content/drive/My Drive/NLP2023/Reviews_Filter.csv')
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
!pwd
```

```python
# Fungsi untuk menghitung compound score
def calculate_sentiment(review):
    sentiment = analyzer.polarity_scores(review)
    return sentiment['compound']

# Hitung compound score untuk setiap review
data['Compound_Score'] = data['Summary'].apply(calculate_sentiment)
```

```python
# Plot distribusi compound score
plt.hist(data['Compound_Score'], bins=30, edgecolor='k')
plt.xlabel('Compound Score')
plt.ylabel('Jumlah Review (N)')
plt.title('Distribusi Compound Score untuk Semua Review')
plt.show()
```

```python
# Filter review yang mengandung kata "coffee"
coffee_reviews = data[data['Summary'].str.contains('coffee', case=False, na=False)]

# Plot compound score untuk review yang mengandung kata "coffee"
plt.hist(coffee_reviews['Compound_Score'], bins=30, edgecolor='k')
plt.xlabel('Compound Score')
plt.ylabel('Jumlah Review (N)')
plt.title('Distribusi Compound Score untuk Review yang Mengandung "Coffee"')
plt.show()
```

```python
# Filter review yang mengandung kata "price"
price_reviews = data[data['Summary'].str.contains('price', case=False, na=False)]

# Plot compound score untuk review yang mengandung kata "price"
plt.hist(price_reviews['Compound_Score'], bins=30, edgecolor='k')
plt.xlabel('Compound Score')
plt.ylabel('Jumlah Review (N)')
plt.title('Distribusi Compound Score untuk Review yang Mengandung "Price"')
plt.show()
```
