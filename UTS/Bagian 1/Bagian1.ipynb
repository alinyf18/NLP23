```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Ganti 'namafile.txt' dengan nama file sebenarnya.
file_path = 'pidato_bungtomo.txt'

# Baca file dan cetak isinya
with open(file_path, 'r') as file:
   isi_berita = file.read()
   print(isi_berita)
```

```python
word_tokens = word_tokenize(isi_berita)
print(word_tokens)
```

```python
stop_words = set(stopwords.words('indonesian'))
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
freq_kata_1 = nltk.FreqDist(word_tokens)
freq_kata_2 = nltk.FreqDist(word_tokens_no_stopwords)
print(freq_kata_1.most_common(10))
print()
print(freq_kata_2.most_common(10))
```

```python
import matplotlib.pyplot as plt
```

```python
freq_kata_1.plot(10)
freq_kata_2.plot(10)
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from collections import Counter

# Baca teks pidato dari file
with open('pidato_bungtomo.txt', 'r', encoding='utf-8') as file:
    pidato = file.read()

# Tokenisasi teks
tokens = word_tokenize(pidato)

# Hapus kata-kata stopwords (kata-kata umum yang tidak memberikan makna khusus)
stop_words = set(stopwords.words('indonesian'))
filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]

# Hitung frekuensi kata-kata
word_freq = Counter(filtered_tokens)

# Ambil 20 kata yang paling sering muncul
common_words = word_freq.most_common(20)

# Siapkan data untuk plot
words, frequencies = zip(*common_words)

# Plot kata-kata yang paling sering muncul
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies)
plt.xlabel('Kata-kata')
plt.ylabel('Frekuensi')
plt.title('10 Kata Paling Sering Muncul dalam Pidato')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
